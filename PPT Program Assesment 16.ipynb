ðŸ’¡ **Question 2**

Given a stack of integers, sort it in ascending order using another temporary stack.

**Examples:**

```
Input: a[] = [1, 1, 2, 3, 4, 2, 1]
Output : [-1, -1, 1, 2, 2, 1, -1]

```


````python
from collections import Counter

def find_nearest_elements_with_greater_frequency(arr):
    frequency_dict = Counter(arr)
    stack = []
    result = []

    for i in range(len(arr) - 1, -1, -1):
        while stack and frequency_dict[stack[-1]] <= frequency_dict[arr[i]]:
            stack.pop()

        if not stack:
            result.append(-1)
        else:
            result.append(stack[-1])

        stack.append(arr[i])

    result.reverse()
    return result

# Test the implementation
arr = [1, 1, 2, 3, 4, 2, 1]

print("Input Array:")
print(arr)

result = find_nearest_elements_with_greater_frequency(arr)

print("\nNearest Elements with Greater Frequency:")
print(result)
````

ðŸ’¡ **Question 2**

Given a stack of integers, sort it in ascending order using another temporary stack.

**Examples:**

Input : [34, 3, 31, 98, 92, 23]
Output : [3, 23, 31, 34, 92, 98]

Input : [3, 5, 1, 4, 2, 8]
Output : [1, 2, 3, 4, 5, 8]

````python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()

    def is_empty(self):
        return len(self.items) == 0

    def top(self):
        if not self.is_empty():
            return self.items[-1]

    def sort_stack(self):
        temp_stack = Stack()

        while not self.is_empty():
            temp = self.pop()

            while not temp_stack.is_empty() and temp_stack.top() > temp:
                self.push(temp_stack.pop())

            temp_stack.push(temp)

        while not temp_stack.is_empty():
            self.push(temp_stack.pop())

    def print_stack(self):
        for item in self.items:
            print(item, end=" ")
        print()

# Test the implementation
stack = Stack()

stack.push(34)
stack.push(3)
stack.push(31)
stack.push(98)
stack.push(92)
stack.push(23)

print("Original Stack:")
stack.print_stack()

stack.sort_stack()

print("\nSorted Stack:")
stack.print_stack()
````

Input Â :Â Stack[] = [1, 2, 3, 4, 5]

Input Â :Â Stack[] = [1, 2, 3, 4, 5, 6]

Output :Â Stack[] = [1, 2, 4, 5, 6]

````python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()

    def is_empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)

def delete_middle(stack, k):
    if stack.is_empty() or k == 0:
        stack.pop()
        return

    temp = stack.pop()
    delete_middle(stack, k - 1)
    stack.push(temp)

def delete_middle_element(stack):
    k = stack.size() // 2 + 1
    delete_middle(stack, k)

# Test the implementation
stack = Stack()

stack.push(1)
stack.push(2)
stack.push(3)
stack.push(4)
stack.push(5)
stack.push(6)

print("Original Stack:")
print(stack.items)

delete_middle_element(stack)

print("\nStack after Deleting Middle Element:")
print(stack.items)
````

**Examples :**

Input :Â Queue[] = { 5, 1, 2, 3, 4 }Â 

Output :Â YesÂ 

Pop the first element of the given Queue 

i.e 5.Â Push 5 into the stack.Â 

Now, pop all the elements of the given Queue and push them toÂ second Queue.Â 

Now, pop element 5 in the stack and push it to the second Queue.Â Â Â 

Input :Â Queue[] = { 5, 1, 2, 6, 3, 4 }Â 

Output :Â NoÂ 

Push 5 to stack.Â 

Pop 1, 2 from given Queue and push it to another Queue.Â 

Pop 6 from given Queue and push to stack.Â 

Pop 3, 4 from given Queue and push to second Queue.Â 

Now, from using any of above operation, we cannot push 5Â into the second Queue because it is below the 6 in the stack.

</aside>

````python
from queue import Queue

def check_queue_arrangement(queue):
    stack = []
    second_queue = Queue()
    expected = 1

    while not queue.empty():
        if queue.queue[0] == expected:
            queue.get()
            expected += 1
        elif stack and stack[-1] == expected:
            stack.pop()
            expected += 1
        elif not stack:
            stack.append(queue.get())
        elif queue.queue[0] < stack[-1]:
            second_queue.put(queue.get())
        else:
            return "No"

    while not second_queue.empty():
        if second_queue.get() != expected:
            return "No"
        expected += 1

    return "Yes"

# Test the implementation
queue1 = Queue()
queue1.put(5)
queue1.put(1)
queue1.put(2)
queue1.put(3)
queue1.put(4)

print("Queue 1:")
print(list(queue1.queue))

result1 = check_queue_arrangement(queue1)

print("\nResult:")
print(result1)

queue2 = Queue()
queue2.put(5)
queue2.put(1)
queue2.put(2)
queue2.put(6)
queue2.put(3)
queue2.put(4)

print("\nQueue 2:")
print(list(queue2.queue))

result2 = check_queue_arrangement(queue2)

print("\nResult:")
print(result2)
````

ðŸ’¡ **Question 5**

Given a number , write a program to reverse this number using stack.

**Examples:**

Input : 365
Output : 563

Input : 6899
Output : 9986

````python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()

    def is_empty(self):
        return len(self.items) == 0

def reverse_number(number):
    number_str = str(number)
    stack = Stack()
    reversed_str = ""

    for char in number_str:
        stack.push(char)

    while not stack.is_empty():
        reversed_str += stack.pop()

    reversed_number = int(reversed_str)
    return reversed_number

# Test the implementation
number1 = 365
print("Number:", number1)

reversed_number1 = reverse_number(number1)
print("Reversed Number:", reversed_number1)

number2 = 6899
print("\nNumber:", number2)

reversed_number2 = reverse_number(number2)
print("Reversed Number:", reversed_number2)
````

ðŸ’¡ **Question 6**

Given an integer k and aÂ **[queue](https://www.geeksforgeeks.org/queue-data-structure/)**Â of integers, The task is to reverse the order of the firstÂ **k**Â elements of the queue, leaving the other elements in the same relative order.

Only following standard operations are allowed on queue.

- **enqueue(x) :**Â Add an item x to rear of queue
- **dequeue() :**Â Remove an item from front of queue
- **size() :**Â Returns number of elements in queue.
- **front() :**Â Finds front item.
</aside>

````python
from queue import Queue

def reverse_k_elements(queue, k):
    if k <= 0 or k > queue.qsize():
        return

    stack = []
    for _ in range(k):
        stack.append(queue.get())

    while stack:
        queue.put(stack.pop())

    for _ in range(queue.qsize() - k):
        queue.put(queue.get())

# Test the implementation
queue = Queue()
queue.put(1)
queue.put(2)
queue.put(3)
queue.put(4)
queue.put(5)

print("Original Queue:")
print(list(queue.queue))

k = 3
reverse_k_elements(queue, k)

print("\nQueue after Reversing the First", k, "Elements:")
print(list(queue.queue))
````

Given a sequence of n strings, the task is to check if any two similar words come together and then destroy each other then print the number of words left in the sequence after this pairwise destruction.

**Examples:**

InputÂ : ab aa aa bcd ab

OutputÂ : 3

*As aa, aa destroys each other so,*

*ab bcd ab is the new sequence.*

Input : Â tom jerry jerry tom

Output :Â 0

*As first both jerry will destroy each other.*

*Then sequence will be tom, tom they will also destroy*

*each other. So, the final sequence doesnâ€™t contain any*

*word.*

</aside>

````python
def count_remaining_words(sequence):
    stack = []

    for word in sequence:
        if stack and stack[-1] == word:
            stack.pop()
        else:
            stack.append(word)

    return len(stack)

# Test the implementation
sequence1 = ['ab', 'aa', 'aa', 'bcd', 'ab']
print("Sequence 1:")
print(sequence1)

remaining_words1 = count_remaining_words(sequence1)
print("\nNumber of Words Left:", remaining_words1)

sequence2 = ['tom', 'jerry', 'jerry', 'tom']
print("\nSequence 2:")
print(sequence2)

remaining_words2 = count_remaining_words(sequence2)
print("\nNumber of Words Left:", remaining_words2)
````

Given an array of integers, the task is to find the maximum absolute difference between the nearest left and the right smaller element of every element in the array.

**Note:**Â If there is no smaller element on right side or left side of any element then we take zero as the smaller element. For example for the leftmost element, the nearest smaller element on the left side is considered as 0. Similarly, for rightmost elements, the smaller element on the right side is considered as 0.

**Examples:**
Input : arr[] = {2, 1, 8}
Output : 1
Left smaller  LS[] {0, 0, 1}
Right smaller RS[] {1, 0, 0}
Maximum Diff of abs(LS[i] - RS[i]) = 1

Input  : arr[] = {2, 4, 8, 7, 7, 9, 3}
Output : 4
Left smaller   LS[] = {0, 2, 4, 4, 4, 7, 2}
Right smaller  RS[] = {0, 3, 7, 3, 3, 3, 0}
Maximum Diff of abs(LS[i] - RS[i]) = 7 - 3 = 4

Input : arr[] = {5, 1, 9, 2, 5, 1, 7}
Output : 1

````python
def max_abs_difference(arr):
    n = len(arr)
    leftStack = []
    rightStack = []
    leftSmaller = [0] * n
    rightSmaller = [0] * n
    maxDiff = float('-inf')

    # Find nearest smaller element on the right side
    for i in range(n):
        while rightStack and arr[rightStack[-1]] > arr[i]:
            rightIndex = rightStack.pop()
            rightSmaller[rightIndex] = arr[i]
        rightStack.append(i)

    # Find nearest smaller element on the left side
    for i in range(n - 1, -1, -1):
        while leftStack and arr[leftStack[-1]] > arr[i]:
            leftIndex = leftStack.pop()
            leftSmaller[leftIndex] = arr[i]
        leftStack.append(i)

    # Calculate maximum absolute difference
    for i in range(n):
        diff = abs(leftSmaller[i] - rightSmaller[i])
        maxDiff = max(maxDiff, diff)

    return maxDiff

# Test the implementation
arr1 = [2, 1, 8]
print("Array 1:")
print(arr1)

maxDiff1 = max_abs_difference(arr1)
print("\nMaximum Absolute Difference:", maxDiff1)

arr2 = [2, 4, 8, 7, 7, 9, 3]
print("\nArray 2:")
print(arr2)

maxDiff2 = max_abs_difference(arr2)
print("\nMaximum Absolute Difference:", maxDiff2)

arr3 = [5, 1, 9, 2, 5, 1, 7]
print("\nArray 3:")
print(arr3)

maxDiff3 = max_abs_difference(arr3)
print("\nMaximum Absolute Difference:", maxDiff3)
````
